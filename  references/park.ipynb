# PARK变换

变换

$$
\mathbf{i}_{dq0} = \mathbf{P}\mathbf{i}_{abc}=
\frac23\begin{bmatrix} 
cos\theta	&cos(\theta-120\degree)	&cos(\theta+120\degree)\\
-sin\theta	&-sin(\theta-120\degree)	&-sin(\theta+120\degree)\\ 
\frac12& \frac12& \frac12
\end{bmatrix}
\begin{bmatrix}
i_a\\i_b\\i_c
\end{bmatrix}
$$

反变换

$$
\mathbf{i}_{abc} = \mathbf{P}^{-1}\mathbf{i}_{abc}=
\begin{bmatrix} 
cos\theta 	&-sin\theta	& 1\\
cos(\theta-120\degree)&-sin(\theta-120\degree)&1\\ 
cos(\theta+120\degree)&-sin(\theta+120\degree)& 1
\end{bmatrix}
\begin{bmatrix}
i_d\\i_q\\i_0
\end{bmatrix}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def getP(theta)->np.array:
    p = np.array([
        [math.cos(theta/180*math.pi),math.cos(theta/180*math.pi - 2/3*math.pi),math.cos(theta/180*math.pi + 2/3*math.pi)],
        [-math.sin(theta/180*math.pi),-math.sin(theta/180*math.pi - 2/3*math.pi),-math.sin(theta/180*math.pi + 2/3*math.pi)],
        [0.5,0.5,0.5]
    ])*2/3
    return p

def showI(i):
    ia,ib,ic = i
    ia=np.r_[0,ia]
    ib=np.r_[0,ib]
    ic=np.r_[0,ic]
    l = len(ia)
    plt.plot(ia,np.zeros((l)))
    plt.scatter(ia,np.zeros((l)))
    x = ib*(-math.sqrt(3)/2)
    y = ib*0.5
    plt.plot(x,y)
    plt.scatter(x,y)
    x = ic*(-math.sqrt(3)/2)
    y = ic*(-0.5)
    plt.plot(x,y)
    plt.scatter(x,y)

    

TEST_NUM = 1
ia = np.random.random(TEST_NUM)-0.5
ib = np.random.random(TEST_NUM)-0.5

ic = -(ia+ib)

I = np.c_[ia,ib,ic].T
P = getP(50)
park = P@I


fig = plt.figure(1)
showI([ia,ib,ic])
showI(park)
plt.show()
```
